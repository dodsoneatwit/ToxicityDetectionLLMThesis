```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data
models = ["GPT-4o", "GPT-3.5-Turbo-1106", "GPT-3.5-Turbo-0125"]
input_costs = [2.50, 1.00, 0.50]  # Cost per 1M tokens
output_costs = [10.00, 2.00, 1.50]

# X-axis positioning
x = np.arange(len(models))

# Create figure and axis
plt.figure(figsize=(8, 5))

# Plot input costs
plt.plot(x, input_costs, marker='o', linestyle='-', label='Input Cost per 1M Tokens')
plt.scatter(x, input_costs, color='blue')

# Plot output costs
plt.plot(x, output_costs, marker='s', linestyle='-', label='Output Cost per 1M Tokens', color='red')
plt.scatter(x, output_costs, color='red')

# Formatting
plt.xticks(x, models)
plt.ylabel("Cost (USD)")
plt.title("GPT Model Costs per 1M Tokens")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
```
